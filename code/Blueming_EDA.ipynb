```python
import pandas as pd
import re
import numpy as np

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../result/Dynamite_YouTubeComment_20020_7_21_14.csv")
```

```python
data
```

```python
def like_int(data):
    temp = []
    for like in data:
        unit = re.compile("[가-힇]+").findall(like)
        if unit == ['만']:
            like = float(like.replace(unit[0], ""))
            like *= 10000
        elif unit == ['천']:
            like = float(like.replace(unit[0], ""))
            like *= 1000
        else:
            like = float(like)
        
        temp.append(like)
            
    return temp
```

```python
data['좋아요 개수'] = like_int(data['좋아요 개수'])
```
