```python
df7_8 = pd.concat([
    pd.read_pickle("/home/mglee/VSCODE/git_folder/YouTubeComment/data/NEWS/NEWS_JULY_0710.pkl"),
    pd.read_pickle("/home/mglee/VSCODE/git_folder/YouTubeComment/data/NEWS/NEWS_JULY_0711_0717.pkl"),
    pd.read_pickle("/home/mglee/VSCODE/git_folder/YouTubeComment/data/NEWS/NEWS_JULY_0717_0731.pkl"),
    pd.read_pickle("/home/mglee/VSCODE/git_folder/YouTubeComment/data/NEWS/NEWS_AUG_0809.pkl"),
    pd.read_pickle("/home/mglee/VSCODE/git_folder/YouTubeComment/data/NEWS/NEWS_AUG_0810_0828.pkl"),
    pd.read_pickle("/home/mglee/VSCODE/git_folder/YouTubeComment/data/NEWS/NEWS_AUG_0828_0831.pkl"),
])
```

```python
lst = [df1, df2, df3_4, df5, df6, df7_8]
```

```python
df = pd.DataFrame()

for item in lst:
    col = ['video_id', 'time', 'channel_id',
       'video_title', 'video_description', 'channel_title', 'view', 'like',
       'commentcnt', 'cap', 'comment', 'comment_id', 're_reply',
       'comment_time', 'captiontext']

    df = pd.concat([df, item[col]])
```

```python
df = df.sort_values('time').reset_index(drop = True)
```

```python
df['ymd'] = df['time'].apply(lambda x : x[:10])
```

```python
df.groupby('ymd').size()
```

```python
import matplotlib.pyplot as plt

parameters = {'xtick.labelsize': 30,
          'axes.titlesize': 30,
          'figure.figsize' : (60, 30),
          'figure.dpi' : 200
          }
plt.rcParams.update(parameters)

titles = df.drop_duplicates('video_id')

xs = titles.groupby('ymd').size().index

plt.plot(titles.groupby('ymd').size())
plt.title('daily News Video count')
plt.xticks(ticks=xs, labels= xs, rotation = 45)
plt.locator_params(axis = 'x', nbins = len(xs) / 5)
```

```python
to_int = ['like', 'view', 'commentcnt']

for item in to_int:

    df[item] = df[item].apply(lambda x : int(x))

print(f"like : \n{df.drop_duplicates('video_id')['like'].quantile([.01, 0.1, .3, .5, .7, .9, .99])}")
print(f"View : \n{df.drop_duplicates('video_id')['view'].quantile([.01, 0.1, .3, .5, .7, .9, .99])}")
print(f"comments : \n{df.drop_duplicates('video_id')['commentcnt'].quantile([.01, 0.1, .3, .5, .7, .9, .99])}")
```

```python
df['view_like'] = df['view'] / df['like']
df['view_comment'] = df['view'] / df['commentcnt']
df['like_comment'] = df['like'] / df['commentcnt']
```
