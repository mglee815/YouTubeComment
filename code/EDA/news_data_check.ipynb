```python
import pandas as pd
```

```python
import os

PATH = "../data/NEWS"

df = pd.DataFrame()
for file in os.listdir(PATH):
    if file[-3:] == 'pkl':
        print(file)
        temp = pd.read_pickle(f"{PATH}/{file}")
        df = pd.concat([df, temp], axis = 0)
    else:
        print(f"{file} PASS")

#df.to_pickle(f"{PATH}/df_7_8.pkl")
```

```python
df
```

```python
df['ymd'] = df['time'].apply(lambda x : x[:10])
```

```python
titles_ymd = df[['video_title', 'ymd']]
titles = df[['video_title']].drop_duplicates()
```

```python
titles_ymd.drop_duplicates().groupby('ymd').size()
```

```python
titles
```

```python
titles_ymd['ymd'] = titles_ymd['ymd'].apply(lambda x : x[5:10])
```

```python
titles_ymd.drop_duplicates().groupby('ymd').size()
```

```python
import matplotlib.pyplot as plt

parameters = {'xtick.labelsize': 30,
          'axes.titlesize': 30,
          'figure.figsize' : (60, 30),
          'figure.dpi' : 200
          }
plt.rcParams.update(parameters)

xs = titles_ymd.drop_duplicates().groupby('ymd').size().index

plt.plot(titles_ymd.drop_duplicates().groupby('ymd').size())
plt.title('daily News comments count')
plt.xticks(ticks=xs, labels= xs, rotation = 45)
plt.locator_params(axis = 'x', nbins = 50)
```
